```python
def createdata(data):
    if data == 1:
        x = np.random.rand(100,5)
        df1 = pd.DataFrame(x, columns= ['A', 'B', 'C', 'D', 'E'])
        
    elif data == 2:
        col = [0, 0, 0, 0, 0]
        r1 = [0, 0, 0, 0, 0]
        r2 = [0, 0, 0, 0, 0]
        r3 = [0, 0, 0, 0, 0]
        r4 = [0, 0, 0, 0, 0]

        print('Enter the column names')
        i=0
        for i in [0,1,2,3,4]:
            col[i] = input()
        print(col)
        
        print('Enter the values for First Row')
        i=0 
        for i in [0,1,2,3,4]:
            r1[i] = int(input())
        print(r1)
        
        print('Enter the values for Second Row')
        i=0 
        for i in [0,1,2,3,4]:
            r2[i] = int(input())
        print(r2)
        
        print('Enter the values for Third Row')
        i=0 
        for i in [0,1,2,3,4]:
            r3[i] = int(input())
        print(r3)
        
        print('Enter the values for Fourth Row')
        i=0 
        for i in [0,1,2,3,4]:
            r4[i] = int(input())
        print(r4)
        
        df1 = pd.DataFrame([r1,r2,r3,r4], columns = col)
        
    elif data == 3:
        file = input('Enter the File name with format: ')
        df1 = pd.read_csv(file)
        df1 = pd.DataFrame(df1)
    
    else:
        print('Please Enter something between 1,2 or 3')
        
    return df1    
```

```python
def plotter(plot):
    if plot == 1:
        finalplot = df1.iplot(kind= 'scatter')
    elif plot ==2:
        finalplot = df1.iplot(kind= 'scatter', mode = 'markers', symbol ='x', colorscale='paired')
    elif plot ==3:
        finalplot = df1.iplot(kind= 'bar')

    elif plot ==4:
        finalplot = df1.iplot(kind= 'hist')

    elif plot ==5:
        finalplot = df1.iplot(kind= 'box')

    elif plot ==6:    
        finalplot = df1.iplot(kind= 'surface')

    else:
        finalplot = print('Select Only between 1 to 6')
    return finalplot
        
        
        
```

```python
def plotter2(plot):
    col = input('Enter the number of columsns you want to plot by selection only 1, 2 or 3 ')
    col = int(col)
    
    if col == 1:
        colm = input('Enter the Column name from the dataframe head; ')
        
        if plot == 1:
            finalplot = df1[colm].iplot(kind= 'scatter')
            
        elif plot ==2:
            finalplot = df1[colm].iplot(kind= 'scatter', mode = 'markers', symbol ='x', colorscale='paired')
            
        elif plot ==3:
            finalplot = df1[colm].iplot(kind= 'bar')

        elif plot ==4:
            finalplot = df1[colm].iplot(kind= 'hist')

        elif plot ==5:
            finalplot = df1[colm].iplot(kind= 'box')

        elif plot ==6 or plot==7:    
#             finalplot = df1[colm].iplot(kind= 'surface')
            finalplot = print('Bubble Plot or SLurface Plot needs more than 1 columns')

        else:
            finalplot = print('Select Only between 1 to 7')
            
            
    if col == 2:
        x = input('Enter the Column 1 name from the dataframe head: ')
        y = input('Enter the Column 2 name from the dataframe head: ')
        
        if plot == 1:
            finalplot = df1[[x,y]].iplot(kind= 'scatter')
            
        elif plot ==2:
            finalplot = df1[[x,y]].iplot(kind= 'scatter', mode = 'markers', symbol ='x', colorscale='paired')
            
        elif plot ==3:
            finalplot = df1[[x, y]].iplot(kind= 'bar')

        elif plot ==4:
            finalplot = df1[[x, y]].iplot(kind= 'hist')

        elif plot ==5:
            finalplot = df1[[x, y]].iplot(kind= 'box')

        elif plot ==6:
            finalplot = df1[[x, y]].iplot(kind= 'surface')
            
        elif plot==7: 
            size  = input('Please Enter the Size column for the bubble plot: ')
            finalplot = df1.iplot(kind= 'bubble', x=x, y=y, size = size)

        else:
            finalplot = print('Select Only between 1 to 7')

    if col == 3:
        x = input('Enter the Column 1 name from the dataframe head: ')
        y = input('Enter the Column 2 name from the dataframe head: ')
        z = input('Enter the Column 2 name from the dataframe head: ')
        
        if plot == 1:
            finalplot = df1[[x,y,z]].iplot(kind= 'scatter')
            
        elif plot ==2:
            finalplot = df1[[x,y,z]].iplot(kind= 'scatter', mode = 'markers', symbol ='x', colorscale='paired')
            
        elif plot ==3:
            finalplot = df1[[x,y,z]].iplot(kind= 'bar')

        elif plot ==4:
            finalplot = df1[[x,y,z]].iplot(kind= 'hist')

        elif plot ==5:
            finalplot = df1[[x,y,z]].iplot(kind= 'box')

        elif plot ==6:
            finalplot = df1[[x,y,z]].iplot(kind= 'surface')
            
        elif plot==7: 
            size  = input('Please Enter the Size column for the bubble plot: ')
            finalplot = df1.iplot(kind= 'bubble',  x=x, y=y, z=z, size = size)

        else:
            finalplot = print('Select Only between 1 to 7')
    else:
        finalplot = print('Please Enter 1, 2 or 3')

    return finalplot
```
